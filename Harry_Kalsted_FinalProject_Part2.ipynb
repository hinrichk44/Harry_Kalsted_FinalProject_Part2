```python
#Here I am importing all the necessary libraries I may need for analysis


import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
plt.style.use('fivethirtyeight')
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Reading in of the dataset

brooklyn = pd.read_csv("brooklyn_sales_map.csv")
```

```python
#What does the dataset look like

brooklyn.head(3)
```

```python
#Getting rid of the first column since it's just added space

brooklyn = brooklyn.drop(brooklyn.columns[0], axis=1)
```

```python
#What are the columns I am working with?

brooklyn.columns
```

```python
#Setting up my rudimentary model
#In my head I thought that 'residential_units','commercial_units','land_sqft', and 'gross_sqft' would all be efficient 
#predictors of sales price


X = brooklyn.loc[:,['residential_units','commercial_units','land_sqft','gross_sqft']]
y = brooklyn.loc[:, 'sale_price']
```

```python
#Here I am making sure I selected the right data

print(X)
```

```python
#Once again just making sure I selected the right data

print(y)
```

```python
#I am settting aside 25% of the data for testing the model

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
#Here I am creating the model
lr = LinearRegression()

#Here I am asking the model to learn a function that predicts `y` from `X`
lr.fit(X_train, y_train)
```

```python
#Here I am evaluating the model
#This is a terrible score. If the range is from 0 to 1, mine is way closer to 0. Essentiall saying I'm not predicting much

lr.score(X_test, y_test)
```

```python
#Here I am comparing the predictions to actual outcomes just to see how off I am

list(zip(lr.predict(X_test), y_test))
```
